# Variables - affectation

La définition d'une variable s'effectue grâce au symbole `=` comme ceci :

```ocaml
a = 8
type(a)
```

On vient ici de créer une **variable** dont :
* le nom est `a`
* le **type** est `int` : c'est un entier relatif
* la **valeur** vaut `8`

Le choix du nom de variable est important, on utilise en règle générale des mots **sans espace** et **non accentués**, utiliser le symbole `_` (underscore) si besoin. Enfin, plusieurs variables peuvent être déclarées simultanément.

Par exemple, pour calculer la distance parcourue en 3 h à une vitesse moyenne de 87 km/h on peut écrire :

```ocaml
duree, km_par_h = 3, 87
duree * km_par_h
```

On peut bien entendu effectuer des calculs utilisant des variables avec les opérations élémentaires `+` `-` `*` `/`, et le calcul de puissances s'effectue avec l'opérateur `**`.

```ocaml
3**2
```

# Chaînes de caractères

Une chaîne de caractères est définie à l'aide de guillemets `"` ou `'` comme par exemple :

```ocaml
chaine = 'Ceci est une chaine de caracteres'
```

Les chaînes de caractères sont de type `str` (string) :

```ocaml
type(chaine)
```

On accède à la longueur d'une chaine de caractères avec la fonction `len` :

```ocaml
len(chaine)
```

On accède au caractère d'indice `i` de la chaîne avec l'instruction`chaine[i]` :

```ocaml
chaine[5]
```

<span style="color:red">**/!\ Attention :**</span> la numérotation commence à 0 !

On peut également partir de la fin :

```ocaml
chaine[-3]
```

L'instruction `s[i:j]` renvoie la **tranche** correspondant aux éléments de la chaîne `s` de l'indice `i` à l'indice `j` **exclu** :

```ocaml
chaine[2:15]
```

On peut aussi omettre l'un des indices pour extraire la liste depuis le début ou bien jusqu'à la fin :

```ocaml
chaine[:10]
```

```ocaml
chaine[10:]
```

L'opérateur `+` permet de concaténer deux chaînes de caractères, tandis que l'instruction `k*chaine` concatène k fois la même chaine de caractères :

```ocaml
une = 'une '
deux = 'deux '
pas = une + deux
pas
```

```ocaml
marche = 10 * pas
marche
```

# Tuples

Un tuple (type `tuple`) est constitué de plusieurs éléments (de types éventuellement différents), séparés d'une virgule, entre parenthèses :

```ocaml
t = (False, 2713, 3.14)
type(t)
```

On peut effectuer exactement le même type d'opérations que sur les chaînes de caractères (longueur, accès aux éléments, tranche, concaténation).

# Listes

Une liste Python (type `list`) est définie par une suite d'éléments (de types éventuellement différents), séparés par des virgules, entre crochets :

```ocaml
liste = [4, True, 'mot']
type(liste)
```

Tout comme pour les chaînes de caractères, on accède à la longueur d'une liste grâce à la fonction `len` et aux éléments de celle-ci avec une instruction du type `liste[i]` :

```ocaml
len(liste)
```

```ocaml
liste[0]
```

```ocaml
liste[-1]
```

**<p style = "color:red">Important :</p>** 
Contrairement aux chaînes de caractères et aux tuple qui sont **non mutables** (ou **immuables**), on peut modifier les éléments d'une liste :

```ocaml
liste[1] = False
liste
```

Des **méthodes** sont associées aux objets `list` :
* `liste.append(e)` ajoute l'élément `e` à la fin de la liste `liste`
* `liste.pop()` retire le dernier élément de la liste `liste`, et le renvoie

```ocaml
liste.pop()
```

```ocaml
liste.append([3, 4, 5])
liste
```

On peut **concaténer** deux listes avec l'opérateur `+`, éventuellement plusieurs fois avec l'opérateur `*` :

```ocaml
concat = liste + liste
concat
```

L'instruction `L[i:j]` renvoie la **tranche** correspondant aux éléments de la liste `L` de l'indice `i` à l'indice `j` **exclu** :

```ocaml
concat[2:5]
```

On peut aussi omettre l'un des indices pour extraire la liste depuis le début ou bien jusqu'à la fin :

```ocaml
concat[:4]
```

```ocaml
concat[4:]
```

Les listes peuvent également être définies par **compréhension**, c'est-à-dire qu'on indique entre crochet une formule permettant de déterminer les différents termes :

```ocaml
pairs = [2*k for k in range(20)]
pairs
```

**Remarque :** on a ici utilisé `range(20)` qui permet d'itérer sur les 20 premiers entiers : de 0 à 19.
* `range(n, m)` correspond aux entiers de n à m **exclu**
* `range(n, m, p)` correspond aux entier de n à m **exclu**, avec un pas p

# Booléens

On peut comparer deux nombres (type int ou float) avec les opérateurs `<`, `>`, `<=`,`>=`.

Pour l'égalité et la différence les opérateurs sont moins intuitifs :
* `a == b` teste l'égalité entre `a` et `b`
* `a != b` teste la différence entre `a` et `b`

Il faut donc être vigilant et ne pas écrire `a = b` pour tester l'égalité, cela aurait pour effet d'affecter la valeur de `b` à `a`.

Le résultat obtenu est de type `bool` : 
* `True` pour Vrai
* `False` pour Faux.

On dipose des opérateurs booléens suivants :
* `not` pour la négation
* `and` pour le ET logique
* `or` pour le OU logique

**Exemple :** (essayez de deviner le résultat avant d'exécuter les cellules)

```ocaml
from math import pi
not 3 < pi
```

```ocaml
4 > 0 and (64982 % 2 == 0)
```

```ocaml
462 <= 12 or (45463254 % 5 == 0)
```

**Remarque :** on dit qu'il s'agit d'une *évaluation paresseuse*.
* `a and b`évalue d'abord `a` : s'il s'agit de `True` (ou équivalent), `b` est renvoyé ; sinon c'est `a` qui est renvoyé
* `a or b`évalue `a` : s'il s'agit de `False` (ou équivalent), `b` est renvoyé ; sinon c'est `a` qui est renvoyé

**Exemple :** (prévoir le résultat avant d'exécuter les cellules)

```ocaml
(144 % 3 == 0) and 7
```

```ocaml
[1, 2] or (2 < 3)
```
