# Fonctions

Une fonction se définie comme ceci :
```python
def ma_fonction(mon_argument):
    insructions
```
On utilisera l'instruction ```return``` pour interrompre l'exécution et renvoyer un résultat.

Par exemple, la fonction ci-dessous renvoie le carré d'un nombre :

````ocaml
def carre(x):
    return x ** 2

carre(16)
````

Pour une fonction à plusieurs arguments on écrit par exemple : `def ma_fonction(arg1, arg2, arg3):`.

# Structure conditionnelle if-elif-else

Lorsqu'on souhaite exécuter des instruction différentes selon une condition à vérifier au préalable, on utilise la structure suivante :
```python
if condition:
    instructions        # si la condition est vérifiée, ce sont ces instructions qui sont exécutées
else:
    autres_instructions        # sinon, ce sont celles-ci
```
<span style="color:red">**/!\ Attention :**</span>
* Les `:` après la condition et le `else` sont obligatoires pour indiquer que des instructions vont suivre
* toutes les instructions doivent être indentées de la même façon

**Remarque :** S'il n'y a pas d'instruction particulière à exécuter dans le cas où la condition ne serait pas vérifiée, le `else` est inutile.

S'il y a plusieurs conditions différentes à vérifier, on utilise `elif` :
```python
if condition_1:
    instructions_1        # si la condition_1 est vérifiée, ce sont ces instructions qui sont exécutées
elif condition_2:
    instructions_2        # si la condition_2 est vérifiée, ce sont ces instructions qui sont exécutées
elif condition_3:
    instructions_3        # si la condition_3 est vérifiée, ce sont ces instructions qui sont exécutées      
else:
    autres_instructions        # sinon, ce sont celles-ci
```

# Boucle bornée (for)

Lorsqu'on souhaite répéter un certain nombre de fois une même séquence d'instructions, on utilise une boucle for, selon la syntaxe suivante :
```python
for i in range(n):        # i parcours les indices de 0 à n-1
    instructions
```
ou bien :
```python
for i in range(n, m):        # i parcours les indices de n à m-1
    instructions
```
**Remarque :** en fait, on peut itérer autrement qu'avec `range` mais ce n'est pas au programme (hors dictionnaires).

````ocaml
#2
def max(liste):
    long = len(liste)
    m = liste[0]
    for i in range(1, long):   # i parcours les entiers de 2 à long - 1
        if liste[i] > m:
            m = liste[i]
    return m
````

````ocaml
max([3, -5, 8, 11, 9, -2])
````

````ocaml
#3
def syracuse(a, n):
    liste = []
    u = a
    for k in range(n):
        u = syrac(u)
        liste.append(u)
    return liste
````

````ocaml
#4
def parite(liste):
    long = len(liste)
    pairs = 0
    for i in range(long):
        if liste[i] % 2 == 0:
            pairs = pairs + 1
    return pairs / long
````

````ocaml
parite([4, 3, 7, 3, 8, 9, 11, 23])
````

````ocaml
#4 bis
def parite(liste):
    pairs = 0
    for elem in liste:    # on peut directement itérer sur les éléments de la liste
        if elem % 2 == 0:
            pairs = pairs + 1
    return pairs / long
````

1. Écrire une fonction `tps_vol` telle que `tps_vols(a)` renvoie le temps de vol de la suite de Syracuse en partant de `a`, c'est-à-dire le nombre d'étapes nécessaires avant de tomber sur le nombre 1.<br>
**Remarque :** selon la conjecture de Syracuse, on finit toujours par tomber sur 1, peu importe le nombre choisi au départ.
1. Importer la fonction randint du module random (`from random import randint`) puis définir une fonction `double_6` telle que `double_6()` simule le lancer de deux dés (non truqués) et renvoie le nombre de lancers qui auront été nécessaires pour faire un double 6.<br>
**Remarque :** `randint(n, m)` génère un entier aléatoirement entre `n` et `m` inclus.
1. Écrire une fonction `lancers` telle que `lancers(n)` renvoie une liste de 13 éléments (la liste est indicée de 0 à 12) correspondant à la simulation suivante :
    * On effectue des lancers de deux dés dont on note la somme obtenue
    * À chaque fois, on met à jour la liste des effectifs de chaque entier (à cette étape, `liste[i]` donne le nombre de `i` obtenus)
    * On s'arrête dès que l'un des nombres a été obtenu `n` fois 
    
    On renvoie la liste correspondant à la **fréquence** de chaque nombre obtenue lors des lancers. Les valeurs seront arrondies au millième (`round(valeur, n)` arrondit `valeur` à `n` chiffres après la virgules).

````ocaml
#1
def tps_vol(a):
    u = a
    tps = 0
    while u != 1:
        u = syrac(u)
        tps = tps + 1
    return tps
````

````ocaml
#2
from random import randint

def double_6():
    nb_lancers = 0
    des = 0
    while des != 12:
        des = randint(1, 6) + randint(1, 6)    # randint(1, 6) simule le lancer d'un dé
        nb_lancers = nb_lancers + 1
    return nb_lancers
````

````ocaml
#3
def lancers(n):
    liste = [0 for i in range(13)]    # on initialise une liste contenant 13 fois la valeur 0
    maxi = 0
    while maxi < n:
        de = randint(1, 6) + randint(1, 6)
        liste[de] = liste[de] + 1
        maxi = max(liste)
    s = 0
    for i in range(13):     # ou bien :    for e in liste: s = s + e
        s = s + liste[i]
    return [round(liste[i] / s, 3) for i in range(13)]

lancers(100000)
````

**Remarque :** Pour faire un `6`, il y a $5$ combinaisons possibles sur $36$ combinaisons au total : $6 = 1 + 5 = 2 + 4 = 3 + 3 = 4 + 2 = 5 + 1$.<br>
La probabilité de faire un `6` est donc égale à $\dfrac5{36} \simeq 0,139$. C'est effectivement très proche du résultat obtenu à l'indice `6` dans la liste précédente.
