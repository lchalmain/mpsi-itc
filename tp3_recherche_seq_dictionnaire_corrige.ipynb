**<span style = "color:purple">Exercice :</span>** 

1. Écrire une fonction `cherche` telle que `cherche (elem, liste)` renvoie `True` si `elem` est présent dans `liste` et `False` sinon.<br> ***On écrira une première version avec `range` puis une seconde sans.***
1. Dans le TP précédent on a déjà écrit une fonction `max`. Écrire une fonction `max2` telle que `max2(liste)` renvoie le deuxième plus grand élément de `liste`.

````ocaml
#1
def cherche(elem, liste):
    long = len(liste)
    for i in range(long):
        if liste[i] == elem:
            return True
    return False
````

````ocaml
cherche(3, [5, 7, -2, 3, 8])
````

````ocaml
cherche(3, [])
````

````ocaml
#1 bis
def cherche_bis(elem, liste):
    for e in liste:
        if e == elem:
            return True
    return False
````

````ocaml
cherche_bis(4, [6, 2, 5])
````

````ocaml
cherche_bis(4, [3, 7, 4, 1])
````

````ocaml
#2
def max2(liste):
    if liste[0] < liste [1]:    # initialisation avec les deux premiers éléments de la liste
        m1 = liste[1]
        m2 = liste[0]
    else :
        m1 = liste[0]
        m2 = liste[1]
    taille = len(liste)
    for i in range(3, taille):
        elem = liste[i]
        if elem > m1:
            m2 = m1
            m1 = elem
        elif elem > m2:    # mais elem <= m1
            m2 = elem
    return m2    # inutile de mettre un else
````

# Dictionnaire

Un dictionnaire Python permet d'associer des valeurs à des clés en adoptant la syntaxe suivante :
```python
dico = {cle1 : val1, cle2 : val2, cle3 : val3, ...}
```
**Remarque :** les clés et les valeurs peuvent avoir des types différents, même entre eux.

L'accès aux valeurs d'un dictionnaire se fait de la même façon que pour accéder aux éléments d'une liste, mais en indiquant une clé plutôt qu'un indice :
```python
dico[cle]
```
**Exemple :**

````ocaml
fiche = {'nom' : 'GREEN', 'prenom' : 'Alice', 'age' : 18, 2004 : True}
````

````ocaml
fiche['prenom']
````

````ocaml
type(fiche)
````

**Remarque :** le type correspondant aux dictionnaires est `dict`.

On peut ajouter des éléments à un dictionnaire en utilisant l'opérateur d'affectation `=`. On peut donc tout à fait construire un dictionnaire en partant d'un dictionnaire vide (`dico = {}` ou bien `dico = dict()`) puis ajouter des éléments.

**Exemple :**

````ocaml
fiche['regime'] = 'vegetarien'
fiche['date_naissance'] = '2004/08/16'
fiche
````

À l'inverse, on peut supprimer une entrée d'un dictionnaire avec la fonction `del` :

````ocaml
del fiche[2004]
fiche
````

# Parcours d'un dictionnaire

Le nombre d'entrées d'un dictionnaire est accessible grâce à la fonction `len` :
```python
len(dico)
```
Pour parcourir un dictionnaire avec une boucle for on pourra le faire directement en parcourant les clés du dictionnaire grâce à la méthode `keys` :
```python
for cle in dico.keys():
```
Ou même directement :
```python
for cle in dico:
```
Si on préfère itérer sur les couples (clé, valeur) on utilisera la méthode `items` :
```python
for asso in dico.items():
```
**Exemple :**

````ocaml
len(fiche)
````

````ocaml
for cle in fiche.keys():
    print(cle)
````

````ocaml
for cle in fiche:    # la méthode keys n'est pas nécessaire ici
    print(cle)
````

````ocaml
for asso in fiche.items():
    print(asso)
````

**Remarque :** on peut *deconstruire* un couple en déclarant par exemple `x, y = couple`. Ex :

````ocaml
for asso in fiche.items():
    cle, valeur = asso
    print(cle, ' : ', valeur)
````

On peut même déconstruire directement dans le for :

````ocaml
for cle, valeur in fiche.items():
    print(cle, ' : ', valeur)
````

# Comptage à l'aide d'un dictionnaire

**<span style = "color:purple">Exercice :</span>** 

1. Écrire une fonction `compte` telle que `compte(liste)` (ou bien `compte(chaine)`) renvoie un dictionnaire comprenant les différents éléments de `liste` (ou bien les différents caractères de `chaine`) comme clés et le nombre d'occurences de chacun comme valeur associée.
1. En déduire une fonction `plus_frequent` qui renvoie l'élément le plus fréquent d'une liste ou d'une chaîne de caractères.
1. Modifier la fonction précédente pour qu'elle ne tienne pas compte des espaces pour les chaînes de caractères et vérifier sur une phrase sélectionnée au hasard que la lettre la plus fréquente dans la langue française est bien la lettre `e`.
<hr>

````ocaml
#1
def compte(struct):
    dico = {}
    for elem in struct:
        if not (elem in dico):
            dico[elem] = 1
        else:
            dico[elem] = dico[elem] + 1
    return dico

compte('phrase contenant 8 fois la lettre e et 9 espaces')
````

````ocaml
#2
def plus_frequent(struct):
    dico = compte(struct)
    maxi = 0
    for cle, valeur in dico.items():
        if valeur > maxi:
            cle_max = cle
            maxi = valeur
    return cle_max

plus_frequent('phrase contenant 8 fois la lettre e et 9 espaces')
````

````ocaml
#3
def lettre_freq(chaine):
    dico = compte(chaine)
    maxi = 0
    for cle, valeur in dico.items():
        if valeur > maxi and cle != ' ':
            cle_max = cle
            maxi = valeur
    return cle_max

lettre_freq('phrase contenant 8 fois la lettre e et 9 espaces')
````
