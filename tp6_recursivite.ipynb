# Définition

On dit qu'un fonction est récursive lorsqu'elle s'appelle elle-même. Par exemple, la fonction suivante est récursive :
```python
def factorielle(n):
    if n == 0:        # cas de base
        return 1
    else:
        return n * factorielle(n - 1)
```
**Remarque :** Lorsqu'on définit une fonction récursive, il est essentiel de bien définir un cas de base, c'est-à-dire un cas sur lequel on finira toujours par tomber après les différents appels successifs.
