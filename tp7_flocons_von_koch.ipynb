# Le module Turtle

Le module `turtle` de Python permet de tracer des figures, comme on le ferait avec un stylo.<br>
La "tortue" (pointe du stylo) est initialement placée à l'origine d'un repère, la "tête" orientée à droite, et se déplace sur le graphique en fonction des instructions qu'on lui donne.

Les principales instructions sont les suivantes :
<table> 
    <tr>
        <th> Instruction </th>
        <th> Effet </th>
    </tr>
    <tr>
        <td> forward(n) ou fd(n) </td>
        <td> avance de n pixels </td>
    </tr>
    <tr>
        <td> back(n) ou bk(n) </td>
        <td> recule de n pixels </td>
    </tr>
    <tr>
        <td> left(n) ou lt(n) </td>
        <td> effectue une rotation à gauche de n degrés </td>
    </tr>
    <tr>
        <td> right(n) ou rt(n) </td>
        <td> effectue une rotation à droite de n degrés </td>
    </tr>
    <tr>
        <td> unod() </td>
        <td> annule la dernière action </td>
    </tr>
    <tr>
        <td> penup() ou up() </td>
        <td> lève le stylo : n'écrit plus lors des déplacements </td>
    </tr>
    <tr>
        <td> pendown() ou down() </td>
        <td> abaisse le stylo : écrit lors des déplacements </td>
    </tr>
    <tr>
        <td> goto(x, y) </td>
        <td> va au point de coordonnées (x, y) </td>
    </tr>
    <tr>
        <td> pensize(n) ou width(n) </td>
        <td> fixe l'épaisseur du trait à n pixels </td>
    </tr>
    <tr>
        <td> color(r, v, b) ou color('red') </td>
        <td> fixe la couleur selon le code RVB ou en rouge </td>
    </tr>
    <tr>
        <td> colormode(255) ou colormode(1) </td>
        <td> code RVB sur une échelle de 0 à 255 ou bien de 0 à 1 </td>
    </tr>
    <tr>
        <td> speed(n) </td>
        <td> modifie la vitesse, sur une échelle de 1 à 10 </td>
    </tr>
    <tr>
        <td> speed(0) </td>
        <td> le plus rapide </td>
    </tr>
    <tr>
        <td> hideturtle() ou ht() </td>
        <td> cache la "tortue" (le curseur) </td>
    </tr>
    <tr>
        <td> showturtle() ou st() </td>
        <td> montre la "tortue" (le curseur) </td>
    </tr>
    <tr>
        <td> clear() </td>
        <td> efface l'écran mais ne modifie pas le stylo </td>
    </tr>
    <tr>
        <td> reset() </td>
        <td> efface, recentre le stylo, réinitialise les paramètres </td>
    </tr>
    <tr>
        <td> pen() </td>
        <td> renvoie l'ensemble des paramètres du stylo </td>
    </tr>
    <tr>
        <td> done() </td>
        <td> démarre la boucle d'événements </td>
    </tr>
</table>

# Les flocons de Von Koch

Un flocon de Von Koch s'obtient en partant d'un triangle équilatéral, puis en modifiant successivement chaque arête de la façon suivante :
* On divise l'arête en 3
* On construit un triangle équilatéral ayant pour base le segment du milieu
* On ôte le segment du milieu
* On effectue de nouveau ces étapes sur chacune des arêtes obtenues
