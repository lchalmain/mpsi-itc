***Indication :*** On pourra utiliser l'instruction `break` qui permet d'interrompre l'exécution d'une boucle.

```ocaml
# Exemple d'utilisation de break :
for n in range(2, 10):
    premier = True    # tant qu'on n'a pas trouvé de diviseur, on considère n comme premier
    for x in range(2, n):
        if n % x == 0:    # dès qu'on trouve un diviseur de n
            print(n, '=', x, '*', n//x)
            premier = False
            break    # on sort de la boucle for
    if premier:
        print(n, 'est un nombre premier')
```
