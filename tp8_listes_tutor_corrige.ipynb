# Modules tutor et rcviz

Dans ce TP nous utiliserons les modules `tutor` et `rcviz`qui permettent respectivement de visualiser les instructions exécutées par un programme et les appels récursifs réalisés lors de l'appel d'une fonction récursive. Pour cela on commence par charger les fonctions nécessaires :

```ocaml
from tutor import tutor
from rcviz import viz
```

* On visualise le déroulement de l'exécution d'un programme en ajoutant l'instruction `tutor()` à la fin.
* On visualise l'arbre des appels récursifs en insérant le marqueur `@viz` avant de définir une fonction récursive puis en utilisant la méthode `callgraph` sur la fonction après avoir effectué un appel à cette fonction. 

**Exemple :**

```ocaml
#visualisation du tri à bulles vu dans le TP4
def echange(i, j, L):
    """echange les elts d'indice i et j dans L"""
    temp = L[i]
    L[i] = L[j]
    L[j] = temp
    
def tri_bulles(L):
    taille = len(L)
    liste = L[:]  #copie de L : ne modifie pas en place
    for i in range(taille):  
        # cherche l'elem mini a placer en position i
        for j in range(i+1, taille):
            if liste[j] < liste[i]:
                # si plus petit on met a la position i 
                echange(i, j, liste)
    return liste

liste = [9, 16, 2, 8, 4, 5]
liste_triee = tri_bulles(liste)
tutor()
```

```ocaml
#visualisation des appels récursifs de la fonction fibo (suite de Fibonacci) du TP6
@viz
def fibo(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fibo(n - 1) + fibo(n - 2)
    
fibo(7)
fibo.callgraph()
```

Par exemple, `separe([8, 5, 11, 10, 12, 18, 3, 2], 8)` renvoie `([5, 3, 2], [8, 11, 10, 12, 18])`.
<hr>

```ocaml
#1
def min_max(listeNotes):
    nb = len(listeNotes)
    mini = listeNotes[0]
    maxi = listeNotes[0]
    for i in range(1, nb):
    #parcours les indices de 1 a nb - 1
        note = listeNotes[i]
        if note < mini:
            mini = note
        elif note > maxi:
            maxi = note
    return mini, maxi

test = min_max([8, 5, 11, 10, 12, 18, 11, 3, 2])

#tutor()
```

```ocaml
#2
def mediane(listeNotes):
    """listeNotes non vide"""
    triee = tri_bulles(listeNotes)
    taille = len(triee)
    if taille % 2 == 1:    # nb impair de notes
        return triee[taille // 2]  # med est la note centrale
    else :    # nb pair de notes
        mid = taille // 2  # med est la demi-somme des deux valeurs centrales
        return (triee[mid - 1] + triee[mid]) / 2
    
mediane([8, 5, 11, 10, 12, 18, 3, 2])
```

```ocaml
#3
def separe(listeNotes, seuil):
    taille = len(listeNotes)
    inf = []
    sup = []
    for i in range(taille):
        note = listeNotes[i]
        if note < seuil:
            inf.append(note)
        else:
            sup.append(note)
    return inf, sup

test_listes = separe([8, 5, 11, 10, 12, 18, 3, 2], 8)

tutor()
```

# Tri fusion
**<p style = "color:purple">Exercice :</p>** 
1. Définir une fonction `fusion` telle que si `liste1` et `liste2` sont deux listes triées, alors `fusion(liste1, liste2)` renvoie la liste triée contenant tous les éléments des deux listes.<br>
Par exemple, `fusion([2, 4, 6], [1, 3, 5, 7, 9])` renvoie `[1, 2, 3, 4, 5, 6, 7, 9]`.
1. Définir une fonction `split` telle que `split(liste)` renvoie un couple de deux listes de tailles égales (à un élément près), la première comportant les premiers éléments de `liste`, la seconde les suivants.<br>
*On pourra utiliser des slices :*
    * `L[i:j]` renvoie la liste constituée des éléments de L, de l'indice i (inclus) à j (exclus).
    * `L[i:]` renvoie la liste constituée des éléments de L, de l'indice i (inclus) à la fin.
    * `L[:j]` renvoie la liste constituée des éléments de L, du début à l'indice j (exclus).
1. Définir une fonction **récursive** `tri_fusion` telle que `tri_fusion` renvoie la liste `liste` triée. Cette fonction utilisera une stratégie "diviser pour régner" :
    * `liste` est séparée en deux listes de tailles égales (à un élément près)
    * chacune des deux sous-listes est triée
    * les deux sous-listes triées sont fusionnées en une liste toujours triée
    
    **On pensera à bien identifier le cas de base, pour garantir la terminaison de la fonction.**<br>
    Voici un exemple d'arbre d'appels récursifs que l'on peut obtenir :
    
<hr>

```ocaml
#1
def fusion(liste1, liste2):
    ind1 = 0
    taille1 = len(liste1)
    ind2 = 0
    taille2 = len(liste2)
    #les listes sont parcourues a partir des indices ind1 et ind2
    fusion_listes = []
    while ind1 < taille1 and ind2 < taille2:
    #l'elem le plus ptt est ajoute a la fusion
        if liste1[ind1] < liste2[ind2]:
            fusion_listes.append(liste1[ind1])
            ind1 = ind1 + 1
        else:
            fusion_listes.append(liste2[ind2])
            ind2 = ind2 + 1
    #a la fin du while
    #au moins une des listes est epuisee
    #reste a ajouter les derniers elem
    while ind1 < taille1:
        fusion_listes.append(liste1[ind1])
        ind1 = ind1 + 1
    while ind2 < taille2:
        fusion_listes.append(liste2[ind2])
        ind2 = ind2 + 1
    return fusion_listes

fusion([2, 4, 6], [1, 3, 5, 7, 9])
#tutor()
```

```ocaml
#2
def split(liste):
    taille = len(liste)
    moitie = taille // 2
    return liste[:moitie], liste[moitie:]

print(split([1, 2, 3, 4, 5]))
split([1])
tutor()
```

```ocaml
#3
@viz
def tri_fusion(liste):
    if len(liste) <= 1:  # cas de base : tri d'une liste contenant un unique element
        return liste
    liste1, liste2 = split(liste)
    tri1 = tri_fusion(liste1)
    tri2 = tri_fusion(liste2)
    return fusion(tri1, tri2)
    
tri_fusion([4, 3, 1, 5, 2, 7, 6])
tri_fusion.callgraph()
```
