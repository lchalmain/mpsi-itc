# Construction du triangle de Pascal

Le triangle de Pascal est une présentation des coefficient binomiaux dans un triangle :

1<br>
1  1<br>
1  2  1<br>
1  3  3  1<br>
1  4  6  4  1<br>
1  5  10 10 5  1<br>
1  6  15 20 15 6 1<br>

Chaque ligne s'obtient grâce à la précédente grâce à la formule de Pascal :
$$\forall k, n \in \mathbb{N}^\star, \text{tels que } k < n,\quad \begin{pmatrix} n\\ k\end{pmatrix}~=~\begin{pmatrix} n-1\\k-1 \end{pmatrix}~+~\begin{pmatrix} n-1\\ k \end{pmatrix}$$
On a en outre, $$\forall n\in \mathbb{N},\quad \begin{pmatrix} n\\ 0\end{pmatrix}~=~\begin{pmatrix} n\\n \end{pmatrix}~=~1$$
Le coefficient binomial $\begin{pmatrix}n\\k\end{pmatrix}$ se lit alors ligne $n$, colonne $k$ (en numérotant à partir de $0$).

**<span style = "color:purple">Exercice :</span>** 
1. Écrire une fonction `pascal` telle que `pascal(n)` renvoie la liste des listes de coefficient binomiaux $\begin{pmatrix}m\\k\end{pmatrix}$ pour $0\leq k\leq m\leq n$.<br>
Par exemple, `pascal(4)` renvoie :
```python
[[1],
 [1, 1],
 [1, 2, 1],
 [1, 3, 3, 1],
 [1, 4, 6, 4, 1]]
```
2. En déduire une fonction `binom` telle que `binom(k, n)` renvoie le coefficient binomial $\begin{pmatrix}n\\k\end{pmatrix}$.
<hr>

````ocaml
#1
def pascal(n):
    t = [];
    for m in range(n+1):
        ligne = []
        for k in range(m+1):
            if k == 0 or k == m:
                ligne.append(1)
            else:
                b1 = t[m-1][k-1]
                b2 = t[m-1][k]
                ligne.append(b1 + b2)
        t.append(ligne)
    return t

pascal(6)
````

````ocaml
#2
def binom(k, n):
    triangle = pascal(n)
    return triangle[n][k]

assert (binom(0, 5) == 1)
assert (binom(4, 4) == 1)
binom(2, 6)
````

# Tri à bulles

Le tri à bulles est une méthode de tri consistant à rechercher et positionner successivement les éléments, du plus petit au plus grand.<br>
L'algorithme en langage naturel, pour trier une liste `L`, est le suivant :
```
Pour chaque indice i de la liste :
    Pour chaque indice j > i :
        Si L[j] < L[i] : échanger les éléments d'indice i et j
```

**<span style = "color:purple">Exercice :</span>** 

1. Écrire une fonction `echange` telle que `echange(i, j, L)` échange les éléments d'indice `i` et `j` dans la liste `L`.
1. Écrire une fonction `tri` telle que `tri(L)` trie les éléments de la liste `L` dans l'ordre croissant.
1. Exprimer, en fonction de la taille `n` de la liste `L`, le nombre de comparaisons effectuées lors de l'appel de `tri(L)`.
<hr>

````ocaml
#1
def echange(i, j, L):
    temp = L[i]
    L[i] = L[j]
    L[j] = temp
    
liste = [0, 1, 2, 3, 4, 5, 6]
echange(0, 4, liste)
liste
````

````ocaml
#2
def tri(L):
    taille = len(L)
    for i in range(taille):
        for j in range(i+1, taille):
            if L[j] < L[i]:
                echange(i, j, L)
        
liste = [9, 16, 2, 8, 4, 5]
tri(liste)
liste
````

# Recherche d'un facteur dans un texte

On dit qu'une chaîne de caractères `motif` est un facteur de la chaîne de caractères `chaine` si les caractères de `motif` apparaissent consécutivement dans `chaine`. Par exemple `'art'` est une facteur de `'particule'`.

**<span style = "color:purple">Exercice :</span>** 

1. Écrire une fonction `fact` telle que `fact(motif, chaine, i)` renvoie `True` si `motif` est un facteur de `chaine` présent à partir de l'indice `i` et `False` sinon. Par exemple, `fact('art', 'particule', 1)` renvoie `True` tandis que `fact('art', 'particule', 4)` renvoie `False`.
1. En déduire une fonction `facteur` telle que `facteur(motif, chaine)` renvoie `True` si `motif` est un facteur de `chaine` et `False` sinon.
<hr>

````ocaml
#1
def fact(motif, chaine, i):
    """On suppose que l'indice i est bien choisi pour qu'il n'y ait pas de dépassement"""
    long = len(motif)
    for j in range(long):
        if motif[j] != chaine[i+j]:
            return False
    return True

assert (fact('art', 'particule', 1))
assert (not (fact('art', 'particule', 4)))
````

````ocaml
#2
def facteur(motif, chaine):
    l_motif = len(motif)
    l_chaine = len(chaine)
    for i in range(l_chaine - l_motif + 1):
        if fact(motif, chaine, i):
            return True
    return False

assert (facteur('art', 'particule'))
assert (not (facteur('test', 'particule')))
````
